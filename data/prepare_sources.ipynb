# Healthcare source preparation

Notebook minh họa cách tạo và kiểm tra hai nguồn dữ liệu dùng cho SSIS. Logic chính được đặt trong `scripts/prepare_sources.py` để có thể tái sử dụng và kiểm tra trên CI.

```python
from pathlib import Path
import subprocess
import sys

data_dir = Path.cwd()
if not (data_dir / 'healthcare_admissions.csv').is_file():
    data_dir = data_dir / 'data'
repo_root = data_dir.parent
subprocess.run(
    [
        sys.executable,
        repo_root / 'scripts' / 'prepare_sources.py',
        '--input', data_dir / 'healthcare_admissions.csv',
        '--output-dir', data_dir,
    ],
    check=True,
)
```

```python
import pandas as pd

patients = pd.read_csv(data_dir / 'patient_source.csv')
operations = pd.read_csv(data_dir / 'operation_source.csv')

print(f'Patients: {len(patients):,}')
print(f'Admissions: {len(operations):,}')
patients.head()
```

```python
assert patients['source_id'].is_unique
assert operations['source_id'].isin(patients['source_id']).all()
operations.describe(include='all').T
```
